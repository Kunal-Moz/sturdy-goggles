## Simple Sine Wave

```python
N = 1024 # number of samples
k = 10.0 # frequency / N
x = np.arange(N) 
y = np.sin( 2. * np.pi * k /N * x ) 
Y = fft(y)  # Our FFT implementation
Ynp = np.fft.fft(y) # Numpy's FFT implementation
Y_abs = abs(Y)
Ynp_abs = abs(Y)

f1 = plt.figure(1)
plt.plot( x, y )
plt.xlabel('x')
plt.ylabel('y')

f2 = plt.figure(2)
plt.plot( x, Y_abs, label="our fft" )
plt.plot( x, Ynp_abs, label="numpy fft")
plt.ylabel('Fourier Component')
plt.xlabel('Spectral Index')
plt.legend()
plt.xlim([0,50])

plt.show()
```

```python
# Period of oscillations : 
T = N/k
print("Period of oscillations : ", T)

f1 = plt.figure(1)
plt.plot( x, y )
plt.xlabel('x')
plt.ylabel('y')
plt.xlim([0,102.4])

# Frequency : 
f = k/N
print("Frequency : ", f)

# f2 = plt.figure(2)
# plt.plot( x/N, Y_abs, label="our fft" )
# plt.ylabel('Fourier Component')
# plt.xlabel('Frequency domain')
# plt.legend()
# plt.xlim([0,0.05])
```

$\textbf{Answer :}$

Although this problem was partially done in the class, this report tries to give a detailed account of the solution. 
The first question that was posed is to find the period of oscillation. So our function is given as : 
$F(x) = sin(\frac{2\pi k x}{N}$)  
Where $N$ is the total number of samples and $k$ denotes the wave number. 
So the frequency is given by :  
$ f = \frac{k}{N}$  
So automatically our time period is given by:  
$T = \frac{1}{f} = \frac{k}{N} = 102.4$ units  
This is verified by showing that the sine function completes a cycle in the given number of steps. 

We can now calculate the frequecy as :  
$ f = \frac{k}{N} = 0.00976Hz$ .
