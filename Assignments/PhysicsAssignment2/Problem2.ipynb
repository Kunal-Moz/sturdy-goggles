We analyse the data from [input CO$_2$ data from Mauna Loa Observatory](https://www.esrl.noaa.gov/gmd/ccgg/trends/) using FFT and observe the trends in the data. 

```python
import matplotlib.pyplot as plt

from fft import *
import numpy as np
import math
```

```python
# Make the plots a bit bigger to see
# NOTE: Must be done in a separate cell
plt.rcParams['figure.dpi'] = 150
```

```python
from read_co2 import read_co2

date,cco2 = read_co2('co2_mm_mlo.txt')
x=date
y = cco2
y_valid = y >= 0.
y = y[y_valid]
x = x[y_valid]

print("Power spectra of REDUCED DATA")
print("Size of data : ", len(x))

y_red =y[0:256]
x_red =x[0:256]

Y = fft(y_red)

Y_abs = abs(Y)


f1 = plt.figure(1)
plt.plot( x_red, y_red )
plt.ylabel('CO${2}$, ppm')
plt.xlabel('Date')

xd = np.arange(len(Y_abs))
f2 = plt.figure(2)
plt.plot( xd, Y_abs )
plt.yscale('log')
plt.ylabel('Fourier Component')

plt.xlabel('Spectral Index')

plt.show()
```

```python

#### instead of truncating, pad with values
window = True

N = len(y)
log2N = math.log(N, 2)
next_pow_of_2 = int(log2N) + 1
if log2N - int(log2N) > 0.0 :  
    print("Need to pad")
    ypads = np.full( 2**( next_pow_of_2) - N, 300.0, dtype=np.double)
    y = np.concatenate( (y, ypads) )
    #### CAREFUL: When you pad, the x axis becomes somewhat "meaningless" for the padded values, 
    #### so typically it is best to just consider it an index
    x = np.arange(len(y))
    N = len(y)
    #### Apply a window to reduce ringing from the 2^n cutoff
    if window : 
        print("Used Hann window")
        pp = (0.5 - 0.5 * np.cos(2*np.pi*x/(N-1)))
        y = y * pp
        print(y)
                
Y = fft(y)
#### Smooth the data in the Fourier domain.
#### Adjust this to change the frequencies to delete (frequencies are removed from maxfreq to N/2
#### and accounts for the Nyquist frequency). 
Y1 = Y
##### Full power spectrum 
ps = fft_power(Y) 
psy = ps

#### Trial max frequency
maxfreqt = 256
Y1[maxfreqt:len(Y)-maxfreqt] = 0.0
#### Get the absolute value and power for plotting
powery1 = fft_power(Y1)

#### Actual value used for later calculation
maxfreq = 50
Y[maxfreq:len(Y)-maxfreq] = 0.0
#### Get the absolute value and power for plotting
Y_abs = abs(Y)
powery = fft_power(Y)
powerx = np.arange(powery.size)

#### Now go back to the frequency domain. 

yfiltered = ifft(Y)
#### To compare the data before and after filtering. 
#### Undoing the windowing
if window:
    for ip in range(len(yfiltered)) :
        if pp[ip] > 0.01:
            yfiltered[ip] = yfiltered[ip]/pp[ip]

#### Undoing the padding of the power spectrum
yfiltered = yfiltered[:-len(ypads)]

yfiltered_abs = abs(yfiltered)
```

$\textbf{Answer :}$

This constitutes the first part of the problem where we need to plot a smoothened power spectrum. So we do that first paddind our data with a scalar bias to ensure that $N=2^n$ where $n \in \mathcal{N}$. In this particular case we chose a value of $300 ppm$ as it is closer to the rest of the values. Furthermore we use the Hann window to reduce the ringing observed in the Fourier transform of the time series. Next we take a Fourier Transform and then cut all values above a chosen maximum frequency. So we chose two frequencies, one as a trial $f_{max1} = 256 Hz$ and other one at $f_{max2} = 50Hz$.  
  
Both the frequency cut-off are similar which was quite obvious. We find that there are 3 peaks one which is a very low frequency peak corresponding to the larger oscillations in climate patterns spanning over millenias (speculatively). We observe two other peak one which is a very high frequency, may be corresponding to yearly fluctuations. Finally we can find another peak in a mid range probably related to effects of solar cycles or something else.  

Following that we take the inverse fft to get the time series back from which we removing the windowing and padding to get the original time series back. The time series or the data is plotted with the original data in the following cells. 


```python

M = len(ps)
#print(M)
window = True
log2M = math.log(M, 2)
#print("log of", N,"to the base 2 :",log2N)
next_pow_of_2 = int(log2M) + 1
if log2M - int(log2M) > 0.0 : 
    print("padded power spetrum")
    pspads = np.full( 2**( next_pow_of_2) - M, 0.0, dtype=np.double)
    ps = np.concatenate( (ps, pspads) )
    #### CAREFUL: When you pad, the x axis becomes somewhat "meaningless" for the padded values, 
    #### so typically it is best to just consider it an index
    M = len(ps)
    x2 = np.arange(M)
    #### Apply a window to reduce ringing from the 2^n cutoff
    if window : 
        ps = ps*(0.5 - 0.5 * np.cos(2*np.pi*x2/(N-1)))
                
            
ps_new = fft(ps)
#print(len(ps_new))
maxfreq2 = 25
ps_new[maxfreq2:len(ps_new)-maxfreq2] = 0.0
#### Get the absolute value and power for plotting

#### Now go back to the frequency domain. 
#### Compare the data before and after filtering. 
psfiltered = ifft(ps_new)

####Undoing the windowing
if window:
    for ip in range(len(psfiltered)) :
        if (0.5 - 0.5 * np.cos(2*np.pi*ip/float(N-1))) != 0.0:
            psfiltered[ip] = psfiltered[ip]/(0.5 - 0.5 * np.cos(2*np.pi*ip/float(N-1)))

#### Undoing the padding of the power spectrum
psfiltered = psfiltered[:-len(pspads)]
psfiltered_abs = abs(psfiltered)
```

$\textbf{Answer :}$  
This was second part of the problem were we had to smoothen the time series by removing the higher frequency jitters in that. We do that by essentially smoothening our padding and windowing our power series, taking the FFT and inverse FFT and undoing the padding and windowing in the reverse order. This gives a much smoother power spectrum which almost has none of the high frequency jitters as seen above.  
We plot both the figures together to show a better picture of what is happening. We find that the high frequecny peaks now appear as smaller bumps and the rest of the spectrum is quite smooth.
