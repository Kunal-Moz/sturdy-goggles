# Problem - 1 : Resistor Cube   
In this problem we need to evaluate the equivalent resistor of a combination resistors arranged as a cube. The analyical solution to the problem has been worked out in the report submitted for this assignment. 

```python
import numpy as np
from scipy import linalg
import matplotlib.pyplot as plt
```

```python
# Make the plots a bit bigger to see
# NOTE: Must be done in a separate cell
plt.rcParams['figure.dpi'] = 100
```

```python
r = 1.0
r1 = r2 = r3 = r4 = r5 = r6 = r7 = r8 = r9 = r10 = r11 = r12 = r
```

## Compute Equivalent Resistance

```python
V0 = 5.0
V = np.zeros((6,1))
V[3] = V0

def make_Res(r1,r2,r3,r4,r5,r6,r7,r8,r9,r10,r11,r12):
    return np.array([
    [0.0, r1, -r2, r4, -r5, 0.0                 ],
    [-r12, r12, -r9, r11, r5+r11+r9, r12        ],
    [-r3, r1+r3+r6, r3,  -r6,  0.0,  -r7        ],
    [0.0, r1, 0.0,  r4+r11, r11, 0.0            ],
    [-r8, r8+r3,  r1+r9+r8+r3, 0.0, -r9, r8     ],
    [0.0, -r10-r6, 0.0, r4+r11+r10+r6, r11, -r10]
    ])

R = make_Res(r1,r2,r3,r4,r5,r6,r7,r8,r9,r10,r11,r12)    
print("R = ")
print(R)
print("V = ")
print(V)
```

```python
i = np.linalg.solve(R, V)

print("i =")
print(i)
```

```python
#### Compute the value of Equivalent Resistor
Req = V0/i.item(0)
print(" Numerical estimate of Equivalent Resistance, Req:", Req )
print(" Theoretical estimate : ", 5.0/6.0)
```

## Changing resistance 

```python
Ry1 = []
Ry2 = []
#### Case - 1 : Changing r4
for rx in np.linspace(0.0,20.0,100):
    #print("Rr")
    r1 = r2 = r3 = r5 = r6 = r7 = r8 = r9 = r10 = r11 = r12 = 1.0
    r4 = rx
    R_mat = make_Res(r1,r2,r3,r4,r5,r6,r7,r8,r9,r10,r11,r12)
    #print(R_mat)
    i = np.linalg.solve(R_mat, V)
    Req = V0/i.item(0)
    #print(Req)
    Ry1.append(Req)


#### Case - 2 : Changing r2
for rx in np.linspace(0.0,20.0,100):
    #print("Rr")
    r1 = r4 = r3 = r5 = r6 = r7 = r8 = r9 = r10 = r11 = r12 = 1.0
    r2 = rx
    R_mat = make_Res(r1,r2,r3,r4,r5,r6,r7,r8,r9,r10,r11,r12)
    #print(R_mat)
    i = np.linalg.solve(R_mat, V)
    Req = V0/i.item(0)
    #print(Req)
    Ry2.append(Req)

    

Rx = np.linspace(1.0,10.0,100)
plt.title("Numerical Result")
plt.plot(Rx,Ry1,label="Changing r4")
plt.plot(Rx,Ry2,label="Cahnging r2")
plt.legend()
plt.xlabel("Value of the Resistor changed, R")
plt.ylabel("Value of Equivalent Resistance, Req")


plt.show()
```

```python
x = np.linspace(0.0,10.0,100)
f1 = (4*x + 1)/(4*x + 2)
f2 = (13*x + 2)/(15*x + 3) 

plt.title("Theoretical Result")
plt.plot(x,f2,label="Changing r4")
plt.plot(x,f1,label="Cahnging r2")
plt.legend()
plt.xlabel("Value of the Resistor changed, R")
plt.ylabel("Value of Equivalent Resistance, Req")


plt.show()
```
