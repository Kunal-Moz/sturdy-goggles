# Computation of $pp \rightarrow Z\rightarrow e^+ e^-$ cross section

Here we are going to compute the production cross section for $pp \rightarrow Z \rightarrow e^+ e^-$ using parton distribution functions from MSTW. This is a simple code that we have modified to use `swig`, but is no longer very actively maintained. 

This makes use of objects that are adapted from [ROOT](root.cern.ch), particularly their [Lorentz Vector class](https://root.cern.ch/doc/master/classTLorentzVector.html). 

This code computes the differential production cross section of pp -> Z -> e+e- using the [MSTW 2008 PDF set](https://mstwpdf.hepforge.org/code/code.html).
It also computes the total production cross section. This can be compared to the data collected by
the CMS experiment at 7 TeV [here](https://hepdata.net/record/ins921788).

## Create incoming $q + \bar{q}$.

````python
# create incoming e+ and e-
qu_i  = qft.Particle( qft.LorentzVector(0,0, 7000.0*0.5,7000.2*0.5),  2);
qbu_i = qft.Particle( qft.LorentzVector(0,0, -7000.0*0.5,7000.2*0.5),  -2);
qd_i  = qft.Particle( qft.LorentzVector(0,0, 7000.0*0.5,7000.2*0.5),  1);
qbd_i = qft.Particle( qft.LorentzVector(0,0, -7000.0*0.5,7000.2*0.5),  -1);
qs_i  = qft.Particle( qft.LorentzVector(0,0, 7000.0*0.5,7000.2*0.5),  -3);
qbs_i = qft.Particle( qft.LorentzVector(0,0, -7000.0*0.5,7000.2*0.5),  3);
em_i   = qft.Particle( qft.LorentzVector(0,0, 7000.0*0.5,7000.2*0.5),  11);
ep_i   = qft.Particle( qft.LorentzVector(0,0,-7000.2*0.5,7000.2*0.5), -11);
````

## Compute differential cross section of $q+ \bar{q} \rightarrow Z \rightarrow e^+ e^-$  
$$
\dfrac{d\sigma}{d\Omega} = \frac{N_c^f}{N_c^i}\frac{1}{256\pi^2}\frac{s}{(s - M_z^2)^2 + \Gamma_z^2}[(L^2 + R^2)(L^{\prime 2} + R^{\prime 2})(1 + cos(\theta)) + (L^2 - R^2)(L^{\prime 2} - R^{\prime 2})2cos(\theta)]
$$  

where $N_c^i = 3$ for the quarks and $N_c^f = 1$ and $L$ and $R$ are right hand left hand components.

````python
dcosTheta = 0.05
cosThetas = np.arange(-1,1,dcosTheta)

sigma_u = 0.
dsigma_domega_u = []

for cosTheta in cosThetas:
    # Produce e+e- --> Z --> e+e-
    ds_dO = qft.dSigmaDOmega( qu_i, qbu_i, 11, cosTheta )
    #print( "%6.2f %10.6e" % (cosTheta, ds_dO))
    # Integral = sum( dsigma/domega) * dcos(theta)
    sigma_u += ds_dO * dcosTheta;
    dsigma_domega_u.append(ds_dO)
print("Sigma = ", sigma_u)

sigma_d = 0.
dsigma_domega_d = []

for cosTheta in cosThetas:
    # Produce e+e- --> Z --> e+e-
    ds_dO = qft.dSigmaDOmega( qd_i, qbd_i, 11, cosTheta )
    #print( "%6.2f %10.6e" % (cosTheta, ds_dO))
    # Integral = sum( dsigma/domega) * dcos(theta)
    sigma_d += ds_dO * dcosTheta;
    dsigma_domega_d.append(ds_dO)
print("Sigma = ", sigma_d)

sigma_s = 0.
dsigma_domega_s = []

for cosTheta in cosThetas:
    # Produce e+e- --> Z --> e+e-
    ds_dO = qft.dSigmaDOmega( qs_i, qbs_i, 11, cosTheta )
#     print( "%6.2f %10.6e" % (cosTheta, ds_dO))
    # Integral = sum( dsigma/domega) * dcos(theta)
    sigma_s += ds_dO * dcosTheta;
    dsigma_domega_s.append(ds_dO)
print("Sigma = ", sigma_s)

plt.plot(cosThetas, dsigma_domega_u)
plt.plot(cosThetas, dsigma_domega_d)
plt.plot(cosThetas, dsigma_domega_s)
plt.show()
````

## Computing Total Cross Section using the parton distribution functions.   

$$ 
\dfrac{d\sigma}{d\Omega} = \sum_{a_1,a_2} \int_0^1 dx_1 dx_2 f_{a_1}^{h1}(x_1,M) f_{a_2}^{h2}(x_2,M)E_Q \frac{d\sigma^{a_1a_2}}{d^3\Omega}(x_1P_1,x_2P_2,M^2)
$$

````python
def dS_Do_up(x):
    q1 = qft.Particle( qft.LorentzVector(0,0, 7000.0*0.5*x[0],7000.2*0.5*x[0]),  2);
    q2 = qft.Particle( qft.LorentzVector(0,0, -7000.0*0.5*x[1],7000.2*0.5*x[1]),  -2);
    
    f1 = pdf.parton(2,x[0],2./3.)
    f2 = pdf.parton(-2,x[1],-2./3.)
    ds_dO = qft.dSigmaDOmega( q1, q2, 11, x[2] )
    e = 7000.0*0.5*x[0] + 7000.0*0.5*x[1]
    return f1*f2*e*ds_dO

def dS_Do_dn(x):
    q1 = qft.Particle( qft.LorentzVector(0,0, 7000.0*0.5*x[0],7000.2*0.5*x[0]),  1);
    q2 = qft.Particle( qft.LorentzVector(0,0, -7000.0*0.5*x[1],7000.2*0.5*x[1]),  -1);
    
    f1 = pdf.parton(1,x[0],-1./3.)
    f2 = pdf.parton(-1,x[1],1./3.)
    ds_dO = qft.dSigmaDOmega( q1, q2, 11, x[2] )
    e = 7000.0*0.5*x[0] + 7000.0*0.5*x[1]
    return f1*f2*e*ds_dO

def dS_Do_s(x):
    q1 = qft.Particle( qft.LorentzVector(0,0, 7000.0*0.5*x[0],7000.2*0.5*x[0]),  3);
    q2 = qft.Particle( qft.LorentzVector(0,0, -7000.0*0.5*x[1],7000.2*0.5*x[1]),  -3);
    
    f1 = pdf.parton(3,x[0],-1./3.)
    f2 = pdf.parton(-3,x[1],1./3.)
    ds_dO = qft.dSigmaDOmega( q1, q2, 11, x[2] )
    e = 7000.0*0.5*x[0] + 7000.0*0.5*x[1]
    return f1*f2*e*ds_dO
````
