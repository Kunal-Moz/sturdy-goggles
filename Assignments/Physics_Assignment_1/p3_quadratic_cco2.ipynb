```python
import numpy as np
import matplotlib.pyplot as plt
from read_co2 import read_co2
#from chi_sq_fit_exp import chi_square_fit
from chi_sq_fit_poly import chi_square_fit_poly
import math
```

```python
# Make the plots a bit bigger to see
# NOTE: Must be done in a separate cell
plt.rcParams['figure.dpi'] = 150
```

```python
dates,cco2 = read_co2('co2_mm_mlo.txt')

n = len(cco2)   # number of data points

if n <= 2 :
    print ('Error! Need at least two data points!')
    exit()
    
err = np.ones(n)
```

```python
# Get the CO2 data


[a0, a1, a2, chi_sq] = chi_square_fit_poly(dates,cco2,err)

# Print out results
print (' Least squares fit of', n, 'data points')
print (' -----------------------------------')
print ("    a2 = {0:6.4f} ".format( a2))
print ("    a1 = {0:6.4f} ".format( a1))
print ("    a0 = {0:3.4f} ".format( a0))
print (" chi-square value : ", chi_sq/(n-3) )

x = np.linspace(1958., 2020.0, 2021)
f = a0 + a1*x + a2*x**2
#print(f)
# Plot it
plt.plot( dates , cco2 )
plt.plot(x, f)

plt.title(r"Atmospheric CO$_{2}$ at Mauna Loa Observatory")
plt.xlabel("Date")
plt.ylabel(r"CO$_{2}$ Concentration), ppm")

plt.show()
```

```python
# Time taken for the CO_2 concentration to reach 50000 ppm : 
cc = 50000.0
t = (-a1 + np.sqrt(a1**2 - 4.0*a2*(a0-cc)))/(2.0*(a2))
# Note : we only take the +ve root as time cannot be negetive in this case. 
print("Date when the air would be toxic : ", int(t))
```
