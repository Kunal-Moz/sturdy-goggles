```python
import numpy as np
import matplotlib.pyplot as plt
from read_co2 import read_co2
from chi_sq_fit_exp import chi_square_fit
import math
```

For fitting an exponential curve we do a semi-log fit where we first take the log of the Y-data and then do the $\chi^2$-fitting on $ln(y_i)$ instead of $y_i$s. Since the data has no systematic uncertianty given

```python
dates,cco2 = read_co2('co2_mm_mlo.txt')

cco2_log = np.log(cco2)
#print(data,cco2_log)
n = len(cco2)   # number of data points

if n <= 2 :
    print ('Error! Need at least two data points!')
    exit()

err_cco2 = np.ones(n)
err = err_cco2/(cco2)
```

```python
# Get the CO2 data


[a0, a1, sigma_a0, sigma_a1, chi_sq] = chi_square_fit(dates, cco2_log, err)

p,cov = np.polyfit(dates,cco2_log,1,cov=True)

# Print out results
print (' Least squares fit of', n, 'data points')
print (' -----------------------------------')
print (" slope   a1 = {0:6.6f} +- {1:6.6f} ".format( a1, sigma_a1))
print (" Intercept     a0 = {0:6.2f} +- {1:6.2f} ".format( a0, sigma_a0))
print (" Numpy's values:           b = {0:6.6f} +- {1:6.6f} ".format( p[0], np.sqrt(cov[0,0])))
print ("                           a = {0:6.6f} +- {1:6.6f} ".format( p[1], np.sqrt(cov[1,1])))
print (" chi-square value : ", chi_sq/(n-2))


x = np.linspace(1958., 2020.0, 2021)
f = a0 + a1 * x
#print(f)
# Plot it
plt.plot( dates , cco2_log )
plt.plot(x, f)
plt.title(r"Atmospheric CO$_{2}$ at Mauna Loa Observatory (Semi-log plot)")
#plt.title(r"Atmospheric CO$_{2}$ at Mauna Loa Observatory")
plt.xlabel("Date")
plt.ylabel(r"Log(CO$_{2}$ Concentration)")
#plt.ylabel(r"CO$_{2}$ Concentration, ppm")

plt.show()
```

```python
# Computing the date at which CO2 concentration reaches 50000 ppm
# intial concentration : a0
# final concentration : log(50000)
t = (np.log(50000) - a0)/a1
print("Date when the air would be toxic : ", int(t))
```
