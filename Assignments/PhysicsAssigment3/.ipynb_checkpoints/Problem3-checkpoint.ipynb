# Yukawa Potential  
Scattering in the Yukawa potential is studied in the problem. Yukawa potential is given as :   
$V(r) = \frac{V_0}{r}e^{-r/r_0}$

```python
import scattering
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Make the plots a bit bigger to see
# NOTE: Must be done in a separate cell
plt.rcParams['figure.dpi'] = 100
```

## Plot potential

```python
V0 = 1.0
r0 = 2.0
Yk = scattering.Yukawa_potential( V0, r0 )

E = 0.705
b_min = -1.0
b_max = 2.0
n_b = 200

rvals = np.linspace(-5,5.0,100)
Ykvals = [ Yk(r) for r in rvals ]
plt.plot(rvals,Ykvals)
plt.xlabel("r")
plt.ylabel("V(r)")
plt.xlim([-5,5])
plt.ylim([-26,26])
plt.show()
```

## Simulate trajectories and differential cross section for $r_0 = 2$

```python

i=0
theta_s = []
fig1 = plt.figure(1)
for b in np.linspace(b_min, b_max, n_b):
    i = i + 1
    xs = scattering.CrossSection_Yukawa_potential( Yk, E, b, 5.0, 100 )
    #Low N hints numerical instability
    deflection = xs.calculate_trajectory()
    traj = np.asarray( xs.get_trajectory() )
    theta_s.append(np.pi - deflection)
    if i%10 == 0:
        x,y = traj[:,0] * np.cos( traj[:,1] ), traj[:,0] * np.sin(traj[:,1])
        plt.plot(x,y, label="%3.1f"%(b))
        plt.xlabel("x")
        plt.ylabel("y")
    #print(deflection)
    
#plt.legend(loc='right')
der = np.zeros(n_b)

theta_s = np.asarray(theta_s)
b=np.linspace(b_min,b_max,n_b)
der[0] = 1.0

for i in range(0,n_b-1):
    der[i+1] = (theta_s[i+1] - theta_s[i])/(b[i+1] - b[i])

print(np.size(b),np.size(der))

scatt_cs = np.zeros(n_b)   
scatt_cs[:] = b[:]/(np.sin(theta_s[:])*np.abs(der[:]))

fig2 = plt.figure(2)
plt.plot(b,scatt_cs,label="Yukawa potential r_0 = 2")
plt.xlabel("Impact parameter (b)")
plt.ylabel("Scattering Cross-section")

cs = (b*np.sin(theta_s/2))/np.sin(theta_s)

plt.legend()


plt.show()
```

## $r_0 \rightarrow \infty$, $V(r) \rightarrow \frac{V_0}{r}$ i.e. Columb potential

```python
r0 = 2000000
Yk = scattering.Yukawa_potential( V0, r0 )
i=0
theta_s = []
fig1 = plt.figure(1)
for b in np.linspace(b_min, b_max, n_b):
    i = i + 1
    xs = scattering.CrossSection_Yukawa_potential( Yk, E, b, 5.0, 100 )
    #Low N hints numerical instability
    deflection = xs.calculate_trajectory()
    traj = np.asarray( xs.get_trajectory() )
    theta_s.append(np.pi - deflection)
    if i%10 == 0:
        x,y = traj[:,0] * np.cos( traj[:,1] ), traj[:,0] * np.sin(traj[:,1])
        plt.plot(x,y, label="%3.1f"%(b))
        plt.xlabel("x")
        plt.ylabel("y")
    #print(deflection)
    
#plt.legend(loc='right')
der = np.zeros(n_b)

theta_s = np.asarray(theta_s)
b=np.linspace(b_min,b_max,n_b)
der[0] = 1.0

for i in range(0,n_b-1):
    der[i+1] = (theta_s[i+1] - theta_s[i])/(b[i+1] - b[i])

print(np.size(b),np.size(der))

scatt_cs = np.zeros(n_b)   
scatt_cs[:] = b[:]/(np.sin(theta_s[:])*np.abs(der[:]))

#### Theoretical value of Columb scattering cross-section 

fig2 = plt.figure(2)
plt.plot(b,scatt_cs,label="Yukawa potential r_0 = 2000000")
#plt.plot(b,dcs_th,label="Columb potential")
plt.xlabel("Impact parameter (b)")
plt.ylabel("Scattering Cross-section")
plt.legend()
plt.show()
```

### Comparison with Columb Scattering

```python
dcs_th = np.power((1/(4*E*np.power(np.sin(theta_s),2))),2)
plt.plot(b,dcs_th,label="Columb potential")
plt.xlabel("Impact parameter (b)")
plt.ylabel("Scattering Cross-section")

plt.legend()
plt.show()
```

## Analysis :   
We plot the various scattered trajectories for increasing values of $r_0$ ($r_0 = 2,20,200,200000$). The trajectories and the differential cross-section looks quite similar in all cases. Though the differential cross-section gets smoother and smoother in each succesive iteration of $r_0$.   
Best part is the $r_0 \rightarrow \infty$ limit gives us the similar form of the theoretical result of Columb scattering
