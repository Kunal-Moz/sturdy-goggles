# Scattering from radial potentials

Here we will use some C++ code to calculate trajectories of particles scattering off of radially symmetric potentials centered at the origin. Using this in bare python is **extremely slow** because there are a lot of iterative loops. Doing this in C++ with the "-O3" compiler setting will allow the compiler to vectorize when possible.

To do that, we will use swig to compute the trajectory. It will output the trajectory as $(r,\phi)$ pairs in a structure like 

`std::vector< std::pair<double,double> >`

We can then read those as `numpy` arrays to plot them in the $(x,y)$ plane.

## Answers :   
We also compute the differntial cross section as:    
$\frac{d\sigma}{d\Omega} = \frac{b}{sin(\theta_s)} \mid \frac{db}{d\theta_s} \mid = \frac{b}{sin(\theta_s)} \mid \frac{\Delta}{\Delta \theta_s} \mid$

Note we compute the derivative in the most basic fashion posible i.e. $\frac{df}{dx} = \frac{f_{i+1} - f_i}{x_{i+1} - x_i}$ as we did not ideally have any direct functional relationship between $\theta_s$ and $b$. In such a case we pad the lower end with a value $1$.

## Simulation for Hard Sphere Scattering

### Plot the potential

```python
b_min = 0.1
b_max = 2.5
n_b = 200
V0 = 1.0
hs = scattering.hard_sphere_potential( V0 )

rvals = np.linspace(0.1,1.5,1000)
hsvals = [ hs(r) for r in rvals ]
plt.plot(rvals,hsvals)
plt.xlabel("r")
plt.ylabel("V(r)")
plt.show()
```

### Simulate the trajectory for $E = 0.8$

```python
#### First value of E : 
E = 0.8 
theta_s = []
i = 0
fig1 = plt.figure(1)
for b in np.linspace(b_min, b_max, n_b):
    i = i + 1
    xs = scattering.CrossSection_hard_sphere_potential( hs, E, b, 5.0, 100 )
    #Low N hints numerical instability
    deflection = xs.calculate_trajectory()
    traj = np.asarray( xs.get_trajectory() )
    theta_s.append(np.pi - deflection)
    if i%10 == 0:
        x,y = traj[:,0] * np.cos( traj[:,1] ), traj[:,0] * np.sin(traj[:,1])
        plt.plot(x,y, label="%3.1f"%(b))
        plt.xlabel("x")
        plt.ylabel("y")
    #print(deflection)
    
    
#plt.legend(loc='right')
der = np.zeros(n_b)
#print(theta_s)
theta_s = np.asarray(theta_s)
b=np.linspace(b_min,b_max,n_b)
der[0] = 1.0
#plt.plot(b,theta_s)
#print(b)
for i in range(0,n_b-1):
    der[i+1] = (theta_s[i+1] - theta_s[i])/(b[i+1] - b[i])
    #der[i+1] = (b[i+1] - b[i])/(theta_s[i+1] - theta_s[i])

print(np.size(b),np.size(der))

scatt_cs = np.zeros(n_b)   
scatt_cs[:] = b[:]/(np.sin(theta_s[:])*np.abs(der[:]))

fig2 = plt.figure(2)
plt.plot(b,scatt_cs)
plt.xlabel("Impact parameter (b)")
plt.ylabel("Scattering Cross-section")

cs = (b*np.sin(theta_s/2))/np.sin(theta_s)

#plt.plot(b,cs)


plt.show()
```

### Simulate trajectories for $E=1.6$ 

```python
#### Second value of E : 
E = 1.6
theta_s = []
fig1 = plt.figure(1)
for b in np.linspace(b_min, b_max, n_b):
    i = i + 1
    xs = scattering.CrossSection_hard_sphere_potential( hs, E, b, 5.0, 100 )
    #Low N hints numerical instability
    deflection = xs.calculate_trajectory()
    traj = np.asarray( xs.get_trajectory() )
    theta_s.append(np.pi - deflection)
    if i%10 == 0:
        x,y = traj[:,0] * np.cos( traj[:,1] ), traj[:,0] * np.sin(traj[:,1])
        plt.plot(x,y, label="%3.1f"%(b))
        plt.xlabel("x")
        plt.ylabel("y")
    #print(deflection)
    
#plt.legend(loc='right')
der = np.zeros(n_b)
#print(theta_s)
theta_s = np.asarray(theta_s)
b=np.linspace(b_min,b_max,n_b)
der[0] = 1.0
#plt.plot(b,theta_s)
#print(b)
for i in range(0,n_b-1):
    der[i+1] = (theta_s[i+1] - theta_s[i])/(b[i+1] - b[i])
    #der[i+1] = (b[i+1] - b[i])/(theta_s[i+1] - theta_s[i])

print(np.size(b),np.size(der))

scatt_cs = np.zeros(n_b)   
scatt_cs[:] = b[:]/(np.sin(theta_s[:])*np.abs(der[:]))

fig2 = plt.figure(2)
plt.plot(b,scatt_cs)
plt.xlabel("Impact parameter (b)")
plt.ylabel("Scattering Cross-section")

cs = (b*np.sin(theta_s/2))/np.sin(theta_s)

#plt.plot(b,cs)

plt.show()
```

## Analysis :   
We plot the trajectory and the plot of scattering cross-section vs the impact parameter for two values of the potential - $E < V_0$ and $E > V_0$. We plotted every $10^{th}$ trajectory for the all $200$ values of the tajectory computed as it was too much to plot for this case.  
We observe that when $E < V_0$ as usual the particles scatter of when $b > r_0 = 1$. In the plot for scattering cross-section we observe the same, however I can not explain the rise after $r_0=1$ and subsequent jumps at other values of $b$. 
When $E > V_0$ the particles at low $b$ penetrate into the potential barrier slightly before scattering off. That's why we see two dips in the plot for differential cross-section. 

## Simulation for Lennard-Jones potential

### Plot the potential 

```python
b_min = 0.6
b_max = 2.0
n_b = 2000
V0 = 1.0
lj = scattering.lennard_jones( V0 )


rvals = np.linspace(0.1,4.0,100)
ljvals = [ lj(r) for r in rvals ]
plt.plot(rvals,ljvals)
plt.xlabel("r")
plt.ylabel("V(r)")
plt.xlim([0,3])
plt.ylim([-1.5,6])
plt.show()
```

### Simulate the trajectory for $E = 0.705$

```python
E = 0.705
theta_s = []
fig1 = plt.figure(1)
for b in np.linspace(b_min, b_max, n_b):
    i = i + 1
    xs = scattering.CrossSection_lennard_jones( lj, E, b, 5.0, 100 )
    #Low N hints numerical instability
    deflection = xs.calculate_trajectory()
    traj = np.asarray( xs.get_trajectory() )
    theta_s.append(np.pi - deflection)
    if i%20 == 0:
        x,y = traj[:,0] * np.cos( traj[:,1] ), traj[:,0] * np.sin(traj[:,1])
        plt.plot(x,y, label="%3.1f"%(b))
        plt.xlabel("x")
        plt.ylabel("y")
    #print(deflection)
    
#plt.legend(loc='right')


der = np.zeros(n_b)
#print(theta_s)
theta_s = np.asarray(theta_s)
b=np.linspace(b_min,b_max,n_b)
der[0] = 1.0
#plt.plot(b,theta_s)
#print(b)
for i in range(0,n_b-1):
    der[i+1] = (theta_s[i+1] - theta_s[i])/(b[i+1] - b[i])
    #der[i+1] = (b[i+1] - b[i])/(theta_s[i+1] - theta_s[i])

print(np.size(b),np.size(der))

scatt_cs = np.zeros(n_b)   
scatt_cs[:] = b[:]/(np.sin(theta_s[:])*np.abs(der[:]))

fig2 = plt.figure(2)
plt.plot(b,scatt_cs)
plt.xlabel("Impact parameter (b)")
plt.ylabel("Scattering Cross-section")

cs = (b*np.sin(theta_s/2))/np.sin(theta_s)

#plt.plot(b,cs)


#plt.show()
plt.show()
```
