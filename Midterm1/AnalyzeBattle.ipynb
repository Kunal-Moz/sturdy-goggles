## Part b

Using the [matplotlib hist](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.hist.html) function, create a single plot showing histograms of the `Attacks` for each character. Label your axes and make a legend. 

```python
## Plot here
#entity = ["Arthas","Fordring","Thrall", "Mograine"]
entity = ["Arthas","Fordring","Thrall", "Mograine","Jaina"]
print("Attack power of each entity")
print(np.transpose(npdata[:,:,0]))
plt.xticks(np.arange(0, 50, step=5))
plt.yticks(np.arange(0,20,step=2))
plt.hist(np.transpose(npdata[:,:,0]), bins=[5,10,15,20,25,30,35,40,45,50] , histtype='bar', label=entity)
plt.legend(prop={'size': 10})
plt.title('Entity : Attack')
plt.ylabel('# of Attacks')
plt.xlabel('Attack power')

#plt.hist(npdata[:,:,1], bins=["Arthas","Fordring","Thrall", "Mograine"])
```
